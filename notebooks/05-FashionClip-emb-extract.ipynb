```python
files = os.listdir(config.data_imgs)
files = [os.path.join(config.data_imgs, file) for file in files]

```

```python
files[10]
```

```python
image_embeddings = fclip.encode_images(files, batch_size=100)
```

```python
# create a dataframe with image embeddings and file names
import pandas as pd
import numpy as np
image_embeddings_norm = image_embeddings/np.linalg.norm(image_embeddings, ord=2, axis=-1, keepdims=True)
df = pd.DataFrame(image_embeddings_norm)
df['article_id'] = files
df['article_id'] = df['article_id'].apply(lambda x: x.split('/')[-1][1:-4])
# parquet must have string column names
df.columns = [str(i) for i in df.columns]
df.to_parquet(os.path.join(config.data_raw_dir, 'image_embeddings_norm.parquet'))
df
```

```python
df = pd.read_parquet(os.path.join(config.data_raw_dir, 'articles.parquet'))
df[['article_id','detail_desc']]
```

```python
# list string columns
df.select_dtypes(include=['object']).columns
# create a new column with all string columns concatenated
df['text'] = df[df.select_dtypes(include=['object']).columns].apply(lambda x: ' '.join(x.dropna().astype(str)), axis=1)
```

```python
text_list = df['text'].to_list()
```

```python
text_embeddings = fclip.encode_text(text_list, batch_size=32)
```

```python
# text_embeddings_norm = text_embeddings/np.linalg.norm(text_embeddings, ord=2, axis=-1, keepdims=True)
# create a dataframe with text embeddings and article ids from the articles dataframe
new_dataframe = pd.DataFrame(text_embeddings)
article_list = df['article_id'].to_list()

new_dataframe['article_id'] = article_list
new_dataframe.columns = [str(i) for i in new_dataframe.columns]
new_dataframe.to_parquet(os.path.join(config.data_raw_dir, 'text_embeddings.parquet'))
new_dataframe
```
